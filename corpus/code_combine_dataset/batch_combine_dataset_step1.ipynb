```python
import pandas as pd
import numpy as np
import random
import os
```

```python
# 定义文件夹路径
input_directory = "D:\\output"
output_directory = "D:\\output"
corpus_path = "D:\\corpusECG\\djy1.txt"
```

```python
# 加载语料库
corpus_df = pd.read_csv(corpus_path)
corpus_dict = {row['Parameter']: eval(row['Values']) for _, row in corpus_df.iterrows()}
```

```python
# 获取文件夹中的所有文件名
file_names = os.listdir(input_directory)
```

```python
# 遍历每个文件
for file_name in file_names:
    # 构造文件的完整路径
    file_path = os.path.join(input_directory, file_name)
    
    # 读取数据文件
    df = pd.read_csv(file_path)
    
    # 提出除'00:00'时刻点其余唯一的时刻点长度
    time_lengths = df[df['Time'] != '00:00'].groupby('Time').size().to_dict()
    
    # 基于特定的语料库和唯一的时刻点长度创建一个新的数据集
    new_data = []
    for time, length in time_lengths.items():
        selected_parameters = np.random.choice(list(corpus_dict.keys()), length)
        for param in selected_parameters:
            value = random.choice(corpus_dict[param])
            new_data.append({'Time': time, 'Parameter': param, 'Value': value})
    
    # 把字典结构转换成DataFrame格式
    new_df = pd.DataFrame(new_data)
    
    # 保留原始数据集的‘00:00’时刻的值并且附加到新的数据集中
    final_df = pd.concat([df[df['Time'] == '00:00'], new_df], ignore_index=True)
    final_df = final_df.sort_values(by='Time').reset_index(drop=True)
    
    # 保存最终数据集文件
    output_file_path = os.path.join(output_directory, file_name)
    final_df.to_csv(output_file_path, index=False, sep=',', header=True)

    print(f"Processed and saved: {output_file_path}")
```
