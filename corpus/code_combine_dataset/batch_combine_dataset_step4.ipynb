```python
import os
```

```python
def move_record_id_to_top(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()  # Read all lines into a list

    # Identify and remove the RecordID line, then re-insert it after the header
    record_id_line = None
    for i, line in enumerate(lines):
        if 'RecordID' in line:
            record_id_line = lines.pop(i)  # Remove the RecordID line
            break

    # Insert the RecordID line right after the header if found
    if record_id_line:
        lines.insert(1, record_id_line)  # Insert after the first line (header)

    # Write the modified content back to the same file to overwrite it
    with open(file_path, 'w') as file:
        file.writelines(lines)
```

```python
def batch_process_files(directory_path):
    # List all files in the given directory
    for filename in os.listdir(directory_path):
        file_path = os.path.join(directory_path, filename)
        
        # Check if it's a file and not a directory
        if os.path.isfile(file_path):
            move_record_id_to_top(file_path)
            print(f"Processed and overwritten: {filename}")
```

```python
# Specify the directory containing your files
directory_path = "D:\\output"  # Update this path to your directory
```

```python
# Call the batch processing function
batch_process_files(directory_path)
```
