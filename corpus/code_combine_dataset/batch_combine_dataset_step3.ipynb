```python
import pandas as pd
import numpy as np
import random
import os
```

```python
# 定义文件夹路径
input_directory = "D:\\output"
output_directory = "D:\\output"
corpus_path = "D:\\corpusECG\\djy1.txt"
```

```python
# 加载语料库
corpus_df = pd.read_csv(corpus_path)
corpus_dict = {row['Parameter']: eval(row['Values']) for _, row in corpus_df.iterrows()}
```

```python
# 2. 定义你希望的参数排序顺序
desired_order = ['RecordID', 'Gender']
```

```python
# 获取文件夹中的所有文件名
file_names = os.listdir(input_directory)
```

```python
# 遍历每个文件
for file_name in file_names:
    # 构造文件的完整路径
    file_path = os.path.join(input_directory, file_name)
    
    # 读取数据文件
    df = pd.read_csv(file_path)
    
    # 3. 过滤出 '00:00' 时刻的数据
    midnight_data = df[df['Time'] == '00:00']
    
    # 4. 保留 'RecordID' 的第一行数据，并确保类型为整数
    record_id_first_row = midnight_data[midnight_data['Parameter'] == 'RecordID'].iloc[0]
    record_id_first_row['Value'] = int(record_id_first_row['Value'])
    
    # 5. 从 corpus1.csv 中获取其余 '00:00' 时刻需要的参数的默认值
    new_data = []
    for param in desired_order[1:]:  # 跳过 RecordID 因为它已经存在
        value = random.choice(corpus_dict[param])
        new_data.append({'Time': '00:00', 'Parameter': param, 'Value': value})
    
    new_df = pd.DataFrame(new_data)
    new_df = pd.concat([new_df, pd.DataFrame([record_id_first_row])], ignore_index=True)
    
    # 7. 对 '00:00' 时刻的数据设置参数排序
    new_df['Parameter'] = pd.Categorical(new_df['Parameter'], categories=desired_order, ordered=True)
    midnight_sorted = new_df.sort_values(by='Parameter')
    
    # 8. 删除原有 '00:00' 时刻数据，合并新的排序数据
    final_df = df[df['Time'] != '00:00']
    final_df = pd.concat([final_df, midnight_sorted], ignore_index=True)
    
    final_df = final_df.sort_values(by=['Time', 'Parameter']).reset_index(drop=True)
    
    # 保存最终数据集文件
    output_file_path = os.path.join(output_directory, file_name)
    final_df.to_csv(output_file_path, index=False, sep=',', header=True)

    print(f"Processed and saved: {output_file_path}")
    
```
